```python
from sklearn.preprocessing import OneHotEncoder
atributoscategoricos = base[['BusinessTravel', 'Department', 'EducationField',
                             'Gender', 'JobRole', 'MaritalStatus', 'OverTime']]
```

```python
dummy = OneHotEncoder()
atributoscategoricos = dummy.fit_transform(atributoscategoricos).toarray()
```

```python
atributoscategoricos
```

```python
baseX = base[['Age', 'DailyRate', 'DistanceFromHome', 'Education', 'EnvironmentSatisfaction',
              'HourlyRate', 'JobInvolvement','JobLevel', 'JobSatisfaction', 'MonthlyIncome', 'MonthlyRate',
              'NumCompaniesWorked', 'PercentSalaryHike', 'PerformanceRating', 'RelationshipSatisfaction',
              'StockOptionLevel', 'TotalWorkingYears', 'TrainingTimesLastYear', 'WorkLifeBalance',
              'YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion', 'YearsWithCurrManager']]
```

```python
atributoscategoricos = pd.DataFrame(atributoscategoricos)
baseprocessada = pd.concat([baseX, atributoscategoricos], axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
normalizar = MinMaxScaler()
baseprocessada.columns = baseprocessada.columns.astype(str)
basenormalizada = normalizar.fit_transform(baseprocessada)
```

```python
baseY = base['Attrition']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(basenormalizada, baseY, test_size=0.3)
```

```python
# Regressão logistica
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

logistic = LogisticRegression()
logistic.fit(X_train, y_train)
```

```python
y_pred = logistic.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier()
forest.fit(X_train, y_train)
```

```python
y_pred = forest.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(hidden_layer_sizes=(18, 18), activation='relu', max_iter=2000,
                    alpha=0.001, solver='sgd', verbose=0, random_state=121)
mlp.fit(X_train, y_train)
```

```python
y_pred = mlp.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
# Padronizar os dados
scaler = StandardScaler()
cartao_padronizado = scaler.fit_transform(cartao_credito)
```

```python
# Aplicar o K-Means
# Método do Cotovelo
wcss_1 = []
range_values = range(1,20)
for i in range_values:
  kmeans = KMeans(n_clusters=i)
  kmeans.fit(cartao_padronizado);
  wcss_1.append(kmeans.inertia_)
```

```python
plt.plot(wcss_1, '-*')
```

```python
kmeans = KMeans(n_clusters=8)
kmeans.fit(cartao_padronizado)
labels = kmeans.labels_
print(labels)
```

```python
np.unique(labels, return_counts=True)
```

```python
kmeans.cluster_centers_
```

```python
cluster_centers = pd.DataFrame(data = kmeans.cluster_centers_,
                               columns = [cartao_credito.columns])
cluster_centers.head()
```

```python
cluster_centers = scaler.inverse_transform(cluster_centers)
cluster_centers = pd.DataFrame(data = cluster_centers,
                               columns = [cartao_credito.columns])
```

```python
cluster_centers
```

```python
cartao_credito_cluster = pd.concat([cartao_credito, pd.DataFrame({'cluster':labels})], axis=1)
```

```python
cartao_credito_cluster
```

```python
for i in cartao_credito.columns:
  plt.figure(figsize=(32, 5))
  for j in range(8):
    plt.subplot(1,8,j+1)
    cluster = cartao_credito_cluster[cartao_credito_cluster['cluster']==j]
    cluster[i].hist(bins=20)
    plt.title('{} Cluster {}'.format(i,j))
plt.show()
```

```python
ordenado = cartao_credito_cluster.sort_values(by='cluster')
ordenado.to_csv('cluster.csv')
```

```python
from prophet import Prophet

def vendas_prediction(id, vendas_df, periodo):
  vendas_df = vendas_df[vendas_df['Store'] == id]
  vendas_df = vendas_df[['Date', 'Sales']].rename(columns={'Date':'ds', 'Sales':'y'})
  vendas_df = vendas_df.sort_values(by='ds')

  model = Prophet()
  model.fit(vendas_df)
  future = model.make_future_dataframe(periods=periodo)
  previsao = model.predict(future)

  figure1 = model.plot(previsao, xlabel = 'Data', ylabel= 'Vendas')
  figure2 = model.plot_components(previsao)

  return vendas_df, previsao
```

```python
vendas_df, previsao = vendas_prediction(10, base_consolidada, 60)
```
